```python
#1. A two-dimensional array consists of a collection of elements organized into rows and columns. 
        #Individual elements are referenced by specifying the specific row and column indices (r, c), both of which start at 0.

class Array2D :

    def __init__(self, numRows, numCols):
        self._Rows = [0]*(numRows)
        for i in range(numRows) :
            self._Rows[i] = [i]*(numCols)

    def numRows(self):
        return len( self._Rows )

    def numCols( self ):
        return len( self._Rows[0] )
    
    def print_matrix( self ):
        print("The matrix is: ")
        for i in range( len(self._Rows )):
            for j in range( len(self._Rows[0]) ):
                print(self._Rows[i][j],end=" ")
            print("")
        return("")

    def clear(self, value):
        print("The matrix is: ")
        i = 0
        while i < len(self._Rows):
            j = 0 
            while j < len(self._Rows[0]):
                self._Rows[i][j] = value
                print(self._Rows[i][j], end = ' ')
                j += 1
            print(" ")
            i += 1
        return ("")

    def getItem(self, i1, i2):
        if i1 < len(self._Rows) and i2 < len(self._Rows[0]):
            print("The value stored at the position is ", end = "")
            return self._Rows[i1][i2]
        else:
            print("The index is not valid")
            return -1

    def setItem(self, i1, i2, value):
        if i1 < len(self._Rows) and i2 < len(self._Rows[0]):
            self._Rows[i1][i2] = value
            print("The number to be changed in the matrix is: ", end = "")
            return self._Rows[i1][i2]
        else:
            print("Please enter valid index.")
            return -1
```

```python
#3. A matrix is a collection of scalar values arranged in rows and columns as a rectangular grid of a fixed size. 
        #The elements of the matrix can be accessed by specifying a given row and column index with indices starting at 0.

class Matrix: 

    def __init__( self, numRows, numCols ):
        self._Rows = [0]*( numRows )
        for i in range(numRows) :
            self._Rows[i] = [0]*( numCols)
    
    def print_matrix( self ):
        print("The matrix is: ")
        for i in range( len(self._Rows )):
            for j in range( len(self._Rows[0]) ):
                print(self._Rows[i][j],end=" ")
            print("")
        print("")

    def setItem(self, i1, i2, value):
        if i1 < len(self._Rows) and i2 < len(self._Rows[0]):
            if value != 0:
                self._Rows[i1][i2] = value
        else:
            print("Please enter valid index.")
            print(-1)
    
    def numRows( self ):
        return len( self._Rows )

    def numCols( self ):
        return len( self._Rows[0] )

    def matrix_to_sparse(self):
        matrix = []
        matrix = matrix + self._Rows
        sparse_matrix=[]
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                if matrix[i][j]!=0:
                    temp=[]
                    temp = temp + [i]
                    temp = temp +[j]
                    temp = temp + [matrix[i][j]]
                    sparse_matrix = sparse_matrix + [temp]
        print("Sparse Matrix:")
        for i in sparse_matrix:
            for j in i:
                print(j,end=' ')
            print()
    
    def getItem(self, i1, i2):
        if i1 < len(self._Rows) and i2 < len(self._Rows[0]):
            print("The value stored at the position is: ", end = "")
            return self._Rows[i1][i2]
        else:
            print("The index is not valid")
            return -1
    
    def scalar(self, value):
        matrix = []
        matrix = matrix + self._Rows
        for i in range(len(self._Rows)):
            for j in range(len(self._Rows[0])):
                matrix[i][j] = matrix[i][j] * value

    def transpose(self):
        matrix = []
        matrix = matrix + self._Rows
        self._Rows = [[matrix[j][i]  for j in range(len(matrix))] for i in range(len(matrix[0]))]
        print("Transpose Matrix:")
        for i in self._Rows:
            for j in i:
                print(j,end=' ')
            print()
    
    def addition(self, mat2):
        matrix = []
        matrix = matrix + self._Rows
        self._Rows = [[matrix[i][j] + mat2[i][j]  for j in range(len(matrix[0]))] for i in range(len(matrix))]
        print("Addition Matrix:")
        for i in self._Rows:
            for j in i:
                print(j,end=' ')
            print()
    
    def subtract(self,mat2):
        matrix = []
        matrix = matrix + self._Rows
        self._Rows = [[matrix[i][j] - mat2[i][j]  for j in range(len(matrix[0]))] for i in range(len(matrix))]
        print("Subtraction Matrix:")
        for i in self._Rows:
            for j in i:
                print(j,end=' ')
            print()
    
    def multiplication(self, mat2):
        matrix = []
        matrix = matrix + self._Rows
        self._Rows=[[0 for col in range(len(mat2[0]))]for row in range(len(matrix))]
        for i in range(len(matrix)):
            for j in range(len(mat2[0])):
                self._Rows[i][j] = 0
                for k in range(len(mat2)):
                    self._Rows[i][j] += matrix[i][k] * mat2[k][j]
        print("Multiplication Matrix:")
        for i in self._Rows:
            for j in i:
                print(j,end=' ')
            print()
```

```python
a3 = Matrix(3,2)
a3.setItem(0,0,5)
a3.setItem(1,1,4)
a3.print_matrix()
print("No. of rows: ",a3.numRows())
print("No. of cols: ",a3.numCols())
print()
a3.matrix_to_sparse()
print()
print(a3.getItem(1,1))
print()
a3.scalar(2)
a3.print_matrix()
print()
a3.matrix_to_sparse()
```

```python
#4. (Game of Life) A life grid is used to represent and store the area in the game of Life that contains organisms. 
        #The grid contains a rectangular grouping of cells of a finite size divided into rows and columns. 
        #The individual cells, which can be alive or dead, are referenced by row and column indices, both of which start at zero.

class LifeGrid:

    def __init__( self, numRows, numCols ):
        self._Rows = [0]*( numRows )
        for i in range(numRows) :
            self._Rows[i] = [0]*( numCols)
    
    def print_grid( self ):
        print("The life grid is: ")
        for i in range( len(self._Rows )):
            for j in range( len(self._Rows[0]) ):
                print(self._Rows[i][j],end=" ")
            print("")
        print("")
    
    def numRows( self ):
        return len( self._Rows )

    def numCols( self ):
        return len( self._Rows[0] )
    
    def configure(self, coordList):
        i = 0
        while i < len(self._Rows):
            j = 0 
            while j < len(self._Rows[0]):
                self._Rows[i][j] = 0
                j += 1
            i += 1
        self._Rows[coordList[0]][coordList[1]] = 1
        return 

    def clear(self):
        print("The life grid is:")
        i = 0
        while i < len(self._Rows):
            j = 0 
            while j < len(self._Rows[0]):
                self._Rows[i][j] = 0
                print(self._Rows[i][j], end = ' ')
                j += 1
            print(" ")
            i += 1
        print("")
    
    def setItem(self, i1, i2):
        if i1 < len(self._Rows) and i2 < len(self._Rows[0]):
                self._Rows[i1][i2] = 1
        else:
            print("Please enter valid index.")
            print(-1)
    
    def isLiveCell(self, row, col):
        for i in range(len(self._Rows)):
            for j in range(len(self._Rows[0])):
                if self._Rows[row][col] == 1:
                    return True
                else:
                    return False
    
    def numLiveNeighbors(self, row, col):
        neighbors = 0
        if self._Rows[row - 1][col - 1] == 1:
            neighbors += 1
        if self._Rows[row - 1][col] == 1:
            neighbors += 1
        if self._Rows[row][col - 1] == 1:
            neighbors += 1
        if row == (len(self._Rows)) - 1:
            if self._Rows[row][col + 1] == 1:
                neighbors += 1
        elif col == (len(self._Rows[0]) - 1):
            if self._Rows[row + 1][col] == 1:
                neighbors += 1
        elif row != (len(self._Rows)) - 1:
            if self._Rows[row][col + 1] == 1:
                neighbors += 1
        elif col != (len(self._Rows[0]) - 1) and row != (len(self._Rows)) - 1:
            if self._Rows[row][col + 1] == 1:
                neighbors += 1
        if col != (len(self._Rows[0]) - 1) and row != (len(self._Rows)) - 1:
            if self._Rows[row + 1][col] == 1:
                neighbors += 1
        elif col != (len(self._Rows[0]) - 1):
            if self._Rows[row + 1][col] == 1:
                neighbors += 1
        elif col != (len(self._Rows[0]) - 1) and row != (len(self._Rows)) - 1:
            if self._Rows[row + 1][col + 1] == 1:
                neighbors += 1
        if neighbors == 2 or neighbors == 3:
            print("Neighboring cells are: ",neighbors)
        else:
            self._Rows[row][col] = 0
            print("Due to improper neihbours the cell will die")
            print("Neighboring cells are: ",neighbors)
```
